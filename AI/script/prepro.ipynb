```python
import numpy as np
import pandas as pd
```

```python
import glob
g = glob.glob("video/*.jsonl")
```

```python
# g = g[1:100]
```

```python
from tqdm import tqdm  # tqdm_notebook as tqdm

desc_list = []
watch_list = []
df_list = []
for p in tqdm(g):
    try:
        df = pd.read_json(p, orient='records', lines=True, encoding='utf8')
        desc_list += list(df["title"])
        watch_list += list(df["watch_num"])
    except:
        pass
```

```python
df_result = pd.DataFrame()
df_result["title"] = desc_list
df_result["watch_num"] = watch_list
```
